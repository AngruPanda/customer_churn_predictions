## Описание данных  и задачи

Оператор связи хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах.

Задача — обучить на этих данных модель для прогноза оттока клиентов.

Оператор предоставляет два основных типа услуг:
1. Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
2. Интернет. Подключение бывает двух типов: через телефонную линию DSL (англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (англ. fiber optic).

Также абонентам доступен ряд услуг:
1. Интернет-безопасность: антивирус (Device Protection) и блокировка опасных сайтов (Online Security);
2. Выделенная линия технической поддержки (Tech Support);
3. Облачное хранилище файлов для резервного копирования данных (Online Backup);
4. Стриминговое телевидение (Streaming TV) и каталог фильмов (Streaming Movies).

За услуги клиенты могут платить ежемесячно или раз в 1–2 года. Доступны различные способы расчёта и возможность получить электронный чек.

Данные хранятся в Sqlite  — СУБД, в которой база данных представлена одним файлом. Она состоит из нескольких таблиц:
* contract — информация о договорах;
* personal — персональные данные клиентов;
* internet — информация об интернет-услугах;
* phone — информация об услугах телефонии.

Таблица contract
* customerID — ID абонента;
* BeginDate — дата начала действия договора;
* EndDate — дата окончания действия договора;
* Type — тип оплаты: раз в год-два или ежемесячно;
* PaperlessBilling — электронный расчётный лист;
* PaymentMethod — тип платежа;
* MonthlyCharges — расходы за месяц;
* TotalCharges — общие расходы абонента.

Таблица personal
* customerID — ID пользователя;
* gender — пол;
* SeniorCitizen — является ли абонент пенсионером;
* Partner — есть ли у абонента супруг или супруга;
* Dependents — есть ли у абонента дети.

Таблица internet
* customerID — ID пользователя;
* InternetService — тип подключения;
* OnlineSecurity — блокировка опасных сайтов;
* OnlineBackup — облачное хранилище файлов для резервного копирования данных;
* DeviceProtection — антивирус;
* TechSupport — выделенная линия технической поддержки;
* StreamingTV — стриминговое телевидение;
* StreamingMovies — каталог фильмов.

Таблица phone
* customerID — ID пользователя;
* MultipleLines — подключение телефона к нескольким линиям одновременно.

Информация о договорах актуальна на 1 февраля 2020.

Можно заметить, что все столбцы с типом данных object, что не совсем корректно, тк присутствуют числовые данные в TotalCharges и MonthlyCharges, исправим это. Также у всех таблиц есть одинаковый столбец и в последней таблице он имеет другое название - поменяем его на правильное.

```python
data_dogovor['TotalCharges'] = pd.to_numeric(data_dogovor['TotalCharges'], errors='coerce').fillna(0)
data_dogovor['MonthlyCharges'] = pd.to_numeric(data_dogovor['MonthlyCharges'], errors='coerce').fillna(0)

data_phone = data_phone.rename(columns = {
    'CustomerId' : 'customerID'
})
```

В данных можно заметить столбец с датой, который будет тяжело анализировать в сыром виде, поэтому преобразуем его в что-то другое - в количество дней сотрудничества с компанияей. В этом деле также нам поможет столбец с целевой переменной, где отображено в какую дату закончилось сотрудничество, и будем делать следующим образом: 

Если в EndDate стоит 'No', то человек продолжает быть клиентом компании и его количество дней сотрудничества равняется текущей дате(2020-01-02) минус дата в BeginDate.

А в другом случае дата в EndDate, когда человек ушел от компании, минус дата в BeginDate.

Ну а также сразу обработаем значения в EndDate, преобразовав их к бинарному виду.

```python
data_dogovor['DaysContract'] = 1

for i in data_dogovor.index:
    if data_dogovor.loc[i, 'EndDate'] != 'No':
        data_dogovor.loc[i, 'DaysContract'] = (
                                                pd.to_datetime(data_dogovor.loc[i, 'EndDate'])
                                               - pd.to_datetime(data_dogovor.loc[i, 'BeginDate'])
                                              ).total_seconds() / (24 * 3600)
    else:
        data_dogovor.loc[i, 'DaysContract'] = (
                                                pd.to_datetime(current_data)
                                               - pd.to_datetime(data_dogovor.loc[i, 'BeginDate'])
                                              ).total_seconds() / (24 * 3600)
        
data_dogovor['EndDate'] = data_dogovor['EndDate'].apply(lambda x: x if x=='No' else "Yes")
```

## Создание моделей и подбор гиперпараметров

Разделим сборную таблицу с отобранными признаками на обучающую и тестовую выборку. Не забываем, что данные у нас не сбалансированные, поэтому стратифицируем выборки указав параметр stratify =merge_table['EndDate'].

```python
X_train, X_test, y_train, y_test = train_test_split(
    merge_table.drop(['customerID',
                      'BeginDate',
                      'EndDate',
                      'gender',
                      'SeniorCitizen',
                      'Dependents',
                      'InternetService',
                     'PaperlessBilling'],
                axis=1),
    merge_table['EndDate'],
    random_state = RANDOM_STATE,
    test_size = TEST_SIZE,
    stratify =merge_table['EndDate'])
```

Начнем настройку пайплайнов. Первым делом создадим ColumnTransformer для преобразования признаков:
* для категориальных признаков используем OneHotEncoder(drop = 'first')
* для числовых признаков используем различные масштабаторы, переберем их в pipeline

```python
cat_cols = merge_table.drop(['customerID', 'BeginDate', 'EndDate',
                  'MonthlyCharges', 'TotalCharges', 'DaysContract',
                  'gender', 'SeniorCitizen','Dependents',
                   'InternetService', 'PaperlessBilling',
                 ], axis=1).columns
```

```python
preprocessing = ColumnTransformer(
    [
        ('ohe', OneHotEncoder(drop = 'first'), cat_cols),
        ('num', MinMaxScaler(), num_col)
    ], 
    remainder='passthrough'
)
```

Первой обучим модель случайного леса, в параметрах укажим веса классов 'balanced', чтобы учесть их не равную пропорцию

```python
pipe_forest = Pipeline(
    [
        ('preprocessor', preprocessing),
        ('models', RandomForestClassifier(class_weight ='balanced', random_state=RANDOM_STATE))
    ]
)
```

Также в самих параметрах подбирем оптимальное число деревьев, максимальную глубину деревьев, минимальное количество делений и листьев.

```python
param_forest = {
    'preprocessor__num': [
        StandardScaler(),
        MinMaxScaler(),
        RobustScaler()
    ],
    'models__n_estimators': [100, 200, 300],
    'models__max_depth': [5, 10, 15, None],
    'models__min_samples_split': [2, 5, 10],
    'models__min_samples_leaf': [1, 2, 4]
}

forest = GridSearchCV(pipe_forest, param_grid=param_forest, cv=5, scoring='roc_auc')
forest.fit(X_train, y_train)
```

```python
pd.DataFrame(forest.cv_results_)[['rank_test_score', 'mean_test_score','params']].sort_values('rank_test_score')
```

Лучшая модель случайного леса набрала около 0.82 по метрике ROC-AUC, что является хорошим результатом! Продолжим обучение моделей и перейдем к моделям бустинга - CatBoostClassifier. В модели также укажим class_weights, но уже с помощью специальной библиотеки.

```python
class_weights = compute_class_weight(
    'balanced', 
    classes=np.unique(y_train), 
    y=y_train
)

pipe_cat = Pipeline(
    [
        ('preprocessor', preprocessing),
        ('models', CatBoostClassifier(class_weights=class_weights.tolist(), logging_level='Silent', random_state=RANDOM_STATE))
    ]
)
```

```python
param_cat = {
    'preprocessor__num': [
        StandardScaler(),
        MinMaxScaler(),
        RobustScaler()
    ],
       'models__iterations': range(400, 601, 10),
       'models__depth': range(2, 10),
    
}


cat = GridSearchCV(pipe_cat, param_grid=param_cat, cv=5, scoring ='roc_auc')

cat.fit(X_train, y_train)
```

```python
pd.DataFrame(cat.cv_results_)[['rank_test_score', 'mean_test_score','params']].sort_values('rank_test_score')
```

Отличный результат! Модель бустинга превзошла результаты модели случайного леса, хоть и заняло это в два раза больше времени, но результаты того стоили. Теперь попробуем обучить и подобрать гиперпараметры модели нейнронной сети, делать это будем с помощью keras tuner.

keras tuner нет ColumnTransformer, поэтому заранее предобработаем все входные данные с помощью preprocessing.

```python
x_train_nn = preprocessing.fit_transform(X_train)
```

Определим архитектуру нейронной сети в которой будем подбирать гиперпараметры с помощью функции. В нейронной сети будем подбирать коэффициент DropOut, Инициализатор весов, Функции активавации на слоях, количество нейроннов и оптимизатор. 

```python
def build_model(hp):
    model = Sequential()

    initializer_choice = hp.Choice('initializer', values=['he_normal', 'glorot_uniform'])
    initializer = HeNormal() if initializer_choice == 'he_normal' else GlorotUniform()
    
    model.add(Dense(units=hp.Int('units_input', 
                                 min_value=64, 
                                 max_value=256, 
                                 step=64),
                    input_shape=(x_train_nn.shape[1],),
                    kernel_initializer=initializer,))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(rate=hp.Float('dropout_input', 
                                   min_value=0.0, 
                                   max_value=0.5, 
                                   step=0.25)))
    
    activation_choice = hp.Choice('activation', values=['relu', 'tanh'])
    
    model.add(Dense(units=hp.Int('units_hidden_1', 
                                 min_value=128, 
                                 max_value=768, 
                                 step=128),
                    kernel_initializer=initializer,))
    model.add(BatchNormalization())
    model.add(Activation(activation_choice))
    model.add(Dropout(rate=hp.Float('dropout_input', 
                                   min_value=0.0, 
                                   max_value=0.5, 
                                   step=0.25)))
    
    model.add(Dense(units=hp.Int('units_hidden_2', 
                                 min_value=128, 
                                 max_value=768, 
                                 step=128),
                    kernel_initializer=initializer,))
    model.add(BatchNormalization())
    model.add(Activation(activation_choice))
    model.add(Dropout(rate=hp.Float('dropout_input', 
                                   min_value=0.0, 
                                   max_value=0.5, 
                                   step=0.25)))
    
    model.add(Dense(units=hp.Int('units_hidden_3', 
                                 min_value=128, 
                                 max_value=768, 
                                 step=128),
                    kernel_initializer=initializer,))
    model.add(BatchNormalization())
    model.add(Activation(activation_choice)) 
    model.add(Dropout(rate=hp.Float('dropout_input', 
                                   min_value=0.0, 
                                   max_value=0.5, 
                                   step=0.25)))
    
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'SGD']),
        loss='binary_crossentropy',
        metrics=['auc'])
    
    return model

tuner = RandomSearch(
    build_model,
    objective='val_auc',
    max_trials=100,
    directory='C:/Users/Aleks/Desktop/test_directory'
    )
```

Посмотрим, что у нас получилось в архитектуре подбора.

```python
tuner.search_space_summary()
```

Все верно, запустим подбор параметров, укажим количество эпох в 50, этого числа должно хватить.

```python
tuner.search(x_train_nn,                                         # Данные для обучения
             y_train.apply(lambda x: 1 if x =='Yes' else 0),     # Правильные ответы
             batch_size=256,                                     # Размер мини-выборки
             epochs=50,                                          # Количество эпох обучения 
             validation_split=0.2,                               # Часть данных, которая будет использоваться для проверки
             )
```

Видим, что результаты бустинга не удалось перебить, поэтому лучшей моделью остается бустинг, который мы будем тестировать и исследовать дальше.

```python
features = cat.best_estimator_['preprocessor'].transform(X_test)
y_true = y_test.apply(lambda x: 1 if x =='Yes' else 0)
y_score = cat.best_estimator_['models'].predict_proba(pd.DataFrame(features))[:, 1]
y_pred = cat.best_estimator_['models'].predict(pd.DataFrame(features))

metric_roc_auc = roc_auc_score(y_true, y_score)
metric_acc = accuracy_score(y_test, y_pred)

print(f'Метрика ROC-AUC для модели бустинга равна {metric_roc_auc}')
print(f'Метрика accuracy для модели бустинга равна {metric_acc}')
```

Результаты подтвердились и это хорошо, значит наша модель бустинга не переобучилась! Построим confusion_matrix для рассмотрения картины более детально.

```python
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', annot_kws={'fontsize': 14})
plt.ylabel('Истинные значения')
plt.xlabel('Предсказанные значения')
plt.title('Матрица ошибок', pad=15)
plt.show()
```

Видно, что наша модель справляется хорошо, но можно сделать лучше. Для нас важно детектировать людей, которое могут уйти, поэтому важно правильно детектировать первый класс и не допускать назначение класса 0, когда в действительности это класс 1. Если мы хотим снизить эти ошибки, то необходимо занизить порог классификации. Посмотрим какая может быть confusion_matrix при разных порогах.

```python
treshhold = [i * 0.01 for i in range(5, 96, 5)]

for tr in treshhold:
    cm = confusion_matrix(y_test, pd.Series(y_score).apply(lambda x: 'Yes' if x >= tr else 'No'))

    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', annot_kws={'fontsize': 14})
    plt.ylabel('Истинные значения')
    plt.xlabel('Предсказанные значения')
    plt.title(f'Матрица ошибок с порогам классификации {tr}', pad=15)
    plt.show()
```

Как можно заметить оптимальным является случай при пороге 0.5, тк ошибка первого рода не превышает верно предсказанный 1 класс и ошибка второго рода не сильно возрастает по сравнению с предыдущими порогами. Добится варианта при котором ошибка второго рода будет нулевой не предоставляется возможным. Посмотрим на графики полноты и точности модели на графике.

```python
precision, recall, thresholds = precision_recall_curve(y_true, y_score)

plt.plot(recall, precision, marker='.', linewidth=2, label='Модель')
plt.xlabel('Полнота (Recall)')
plt.ylabel('Точность (Precision)')
plt.title('Кривая Precision-Recall')
plt.grid(True, alpha=0.3)

f1_scores = 2 * (precision * recall) / (precision + recall + 1e-8)
best_idx = np.argmax(f1_scores)
plt.scatter(recall[best_idx], precision[best_idx], 
           color='red', s=100, label=f'Лучшая точка (F1={f1_scores[best_idx]:.2f})')

plt.legend()

plt.tight_layout()
plt.show()
```

### Промежуточные выводы

- проведена проверка модели на тестовой выборке - результаты подтвердились
- построена confusion matrix при разных порогах - оптимальный порог 0.5/0.6

Проведем анализ методом SHAP модели бустинга для выявления наиболее влиятельных признаков в модели. Первоначально выделим все названия признаков, тк ColumnTransformer не совсем хочет работать и не удается взять из него имена признаков.

```python
preprocessor = cat.best_estimator_['preprocessor']
feature_names = []

for name, transformer, columns in preprocessor.transformers_:
    if transformer == 'drop':
        continue
    if hasattr(transformer, 'get_feature_names'):
        if callable(getattr(transformer, 'get_feature_names')):
            trans_names = transformer.get_feature_names(columns)
        else:
            trans_names = columns
    else:
        trans_names = columns
    feature_names.extend(trans_names)
```

Теперь преобразуем X_train с помощью preprocessor, создадим TreeExplainer и получим SHAP-value

```python
X_train_inter = cat.best_estimator_['preprocessor'].fit_transform(X_train)
X_train_inter = pd.DataFrame(X_train_inter, columns=feature_names)

X_test_inter = cat.best_estimator_['preprocessor'].transform(X_test)
X_test_inter = pd.DataFrame(X_test_inter, columns=feature_names)

explainer = shap.TreeExplainer(cat.best_estimator_['models'])

shap_values = explainer.shap_values(X_test_inter)
```
